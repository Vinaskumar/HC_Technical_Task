```python
# Load CSV files into DataFrames
persons_data = pd.read_csv('persons_data.csv')
occupancy_data = pd.read_csv('occupancy_data.csv')
majors_data = pd.read_csv('majors_data.csv')
inventory_data = pd.read_csv('inventory_data.csv')
```

```python
# Function to map majors to their IDs
def map_majors_to_ids(majors, majors_data):
    if pd.isnull(majors):
        return ''
    major_list = [major.strip() for major in majors.split(',')]
    major_ids = majors_data[majors_data['name'].isin(major_list)]['id'].tolist()
    return ','.join(map(str, major_ids))

# Map major names to their IDs
persons_data['majorIds'] = persons_data['majors'].apply(lambda x: map_majors_to_ids(x, majors_data))
```

The map_majors_to_ids function splits the comma-separated majors, matches them with majors_data[name], and collects the corresponding id values.
These id values are then joined into a comma-separated string and assigned to the majorIds column.

```python
# Merge bed ID from inventory data
occupancy_data = occupancy_data.merge(inventory_data[['buildingName', 'roomName', 'bedName', 'bedId']], on=['buildingName', 'roomName', 'bedName'], how='left')

# Merge person data with occupancy data
final_data = persons_data.merge(occupancy_data[['personId', 'bedId']], on='personId', how='left')
```
